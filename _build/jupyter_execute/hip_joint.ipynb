# Hip joint 


## Anatomy
The hip joint is a crucial articulation that connects the lower extremities to the axial skeleton, enabling a wide range of motion while providing remarkable stability and supporting the weight of the head, arms, and trunk (HAT). It is a **true ball-and-socket, diarthrodial (synovial) joint**, meaning it has a joint cavity, articular cartilage-covered surfaces, a synovial membrane producing synovial fluid, and a ligamentous capsule.

### Bony Anatomy
The hip joint is formed by the articulation of two main bones: the pelvis (specifically, the acetabulum) and the thigh bone (femur).

*   **Acetabulum**:
    *   This is the **cup-shaped socket** of the hip joint, located on the lateral aspect of the pelvic bone.
    *   It is formed by the fusion of three parts of the **innominate bone** (pelvis): the **ilium** (approximately 40%), **ischium** (40%), and **pubis** (20%). In skeletally immature individuals, these bones are separated by triradiate cartilage, which typically fuses between 14 and 23 years of age.
    *   The acetabulum covers about 170° of the femoral head.
    *   Its actual articular surface appears **lunate-shaped** (horseshoe-shaped) when viewed from inside. The central inferior acetabular fossa within this lunate area is a fat-filled space that houses a synovial-covered fat pad and contains the acetabular attachment of the ligamentum teres.
    *   The socket is completed inferiorly by the **transverse acetabular ligament**.
    *   The acetabulum is angled obliquely in an **anterior, lateral, and inferior direction**. Normally, it is oriented in 15 to 20 degrees of anteversion and 40 degrees of abduction.

*   **Femur**:
    *   The femur is the **longest bone in the body** and its primary functions are weight-bearing and contributing to gait stability.
    *   **Femoral Head**: This is the **rounded ball** at the top of the thigh bone that fits into the acetabulum. It forms approximately **two-thirds of a sphere**.
    *   **Femoral Neck**: This connects the femoral head to the shaft. Its structure plays a role in hip joint function, particularly its **angular relationships** with the femoral shaft, defined by the angle of inclination and the torsion angle. The femoral neck is intracapsular, while the greater and lesser trochanters are extracapsular.
    *   **Trochanters**: Close to the femoral head are two bony parts where muscles are attached: the **greater trochanter** (on the side of the thighbone, serving as a muscle attachment point) and the **lesser trochanter**.

### Cartilage and Labrum
The articulating surfaces of the hip joint are covered by specialized cartilage and a fibrocartilaginous rim.

*   **Articular Cartilage**:
    *   The surfaces of both the **femoral head and the acetabulum are covered with articular (hyaline) cartilage**.
    *   This smooth tissue **cushions the ends of the bones, enables smooth movement**, and acts as a shock absorber.
    *   The thickness of articular cartilage in the hip joint typically ranges from **2 to 4 mm**. It is thickest on the medial-central surface of the femoral head and in the superior-anterior region of the acetabular dome.
    *   The cartilage is **viscoelastic**, meaning its load-bearing pattern can change based on the magnitude of the applied load. Under low loads, joint surfaces are incongruous, but under heavy loads, they become congruous, providing maximum surface contact.

*   **Acetabular Labrum**:
    *   The labrum is a **fibrocartilaginous structure that outlines the acetabular socket**, running around its circumference.
    *   It is a **continuous, usually triangular structure** that attaches directly to the bony rim of the acetabulum. Variants like round, irregular, and flattened shapes have been observed in people without hip pain, with the incidence of triangular labra decreasing with age as rounded and irregular margins increase.
    *   The labrum helps to **deepen the socket and increase its concavity**, enhancing joint stability and helping the joint fit better.
    *   It also forms a **seal for the central compartment** by maintaining negative intra-articular pressure, which resists distraction of the femoral head from the socket and enhances nutrient fluid flow to the articular cartilage.
    *   The labrum is typically **2 to 3 mm thick** and is wider and thinner anteriorly, and thicker posteriorly.
    *   It is thought that the **anterior and superior aspects of the labrum are the most innervated portions**, containing free nerve endings and sensory nerve end organs that produce pain, pressure, and deep sensation.
    *   The **vascular supply** to the labrum is provided by the obturator, superior gluteal, and inferior gluteal arteries. This blood supply is believed to be mostly peripheral, with the outer one-third of the acetabular base being vascular, while the majority is avascular. Labral tears are most likely to occur at the junction of the labrum and articular cartilage, an area termed the "watershed region".

### Ligaments and Joint Capsule
The hip joint is enveloped by a strong capsule reinforced by several powerful ligaments that contribute significantly to its stability.

*   **Joint Capsule**:
    *   The hip joint capsule is **strong and dense**.
    *   It is formed by an intertwining of three separate entities (ligaments).
    *   The capsule is **tight in extension and more relaxed in flexion**.
    *   It is thickened **anterosuperiorly**, where predominant stresses occur, and is relatively thin and loosely attached posteroinferiorly.
    *   The capsule covers the femoral head and neck, attaching to the entire periphery of the acetabulum beyond the labrum and to the intertrochanteric line anteriorly and the femoral neck posteriorly.

*   **Capsular Ligaments**: These ligaments reinforce the joint capsule and help restrain motion. They are coiled around the femoral neck, becoming **tightest in combined extension and medial rotation** and loosest in flexion and lateral rotation.
    *   **Iliofemoral Ligament (Y ligament of Bigelow)**: This is considered the **strongest ligament in the body**. It is located anteriorly, spanning from the anterior inferior iliac spine (AIIS) to the intertrochanteric line of the femur in an inverted "Y" shape. Its primary function is to **prevent excessive extension and hyperextension**, and it plays a significant role in stabilizing upright posture and limiting anterior translation of the hip.
    *   **Pubofemoral Ligament**: Located anteriorly, this ligament arises from the pubic ramus and passes to the anterior surface of the intertrochanteric fossa. It helps prevent **excessive abduction and extension**.
    *   **Ischiofemoral Ligament**: This is the **posterior capsular ligament**, attaching to the posterior surface of the acetabular rim and labrum. It winds tightly on extension, helping to stabilize the hip in extension and preventing excess extension. It also limits medial rotation.
    *   All three capsular ligaments (iliofemoral, pubofemoral, and ischiofemoral) **limit medial rotation** of the femur.

*   **Ligamentum Teres (Ligament of the Head of the Femur)**: This is an intra-articular but extrasynovial accessory joint structure. It is a triangular band attached at one end to both sides of the peripheral edge of the acetabular notch, passing under the transverse acetabular ligament, and attaching at its other end to the fovea of the femoral head. It contributes little to hip stability but some propose it plays a role in joint nutrition, especially in childhood.
*   **Zona Orbicularis (Angular Ligament)**: This ligament encircles the femoral neck like a buttonhole and also plays little role in stability.

### Neurovascular Anatomy
The hip joint has a specific innervation and blood supply.

*   **Nerves**: The hip joint receives innervation from several nerves.
    *   **Femoral Nerve**: Supplies the anterior aspect of the joint and the anterior capsule.
    *   **Obturator Nerve**: Supplies the anteromedial aspect of the joint and the anterior capsule.
    *   **Superior Gluteal Nerve**: Supplies the posterior aspect laterally.
    *   **Nerve to Quadratus Femoris**: Supplies the posterior capsule.
    *   Articular branches from the sciatic nerve also contribute to the posterior aspect.
    *   The anterior capsule and the superior labrum are considered the primary pain generators of the hip joint due to their higher density of nociceptors and mechanoreceptors.
    *   The **lateral femoral cutaneous nerve** is also located near the hip and is prone to damage by inappropriately placed arthroscopic portals.

*   **Blood Supply**: The blood supply to the hip is provided by various arteries, with some variations.
    *   **Medial Circumflex Femoral Artery**: This is a major source of blood supply, along with the lateral circumflex femoral artery, both branching from the profunda femoris (deep artery of the thigh).
    *   **Lateral Circumflex Femoral Artery**: Also a major contributor, branching from the profunda femoris.
    *   **Obturator Artery**: Its posterior division gives rise to the foveal artery (artery to the head of the femur), which travels within the ligamentum teres and helps prevent avascular necrosis if the circumflex arteries are disrupted.
    *   **Superior Gluteal Artery** and **Inferior Gluteal Artery**: Contribute to the vascular supply of the labrum and surrounding buttock/thigh region.
    *   There are two significant anastomoses: the **cruciate anastomosis** (supports the upper thigh) and the **trochanteric anastomosis** (supports the head of the femur).
    *   The **retinacular blood vessels**, arising from a vascular ring at the base of the femoral neck, are the major source of nutrition to the femoral head and neck.

### Muscular Anatomy
The hip joint is surrounded by a large number of powerful and well-balanced muscles that enable a wide range of motion in multiple planes while providing essential stability. There are approximately **22 to 27 muscles** acting on the hip joint.

*   **Grouping of Muscles**: These muscles can be divided into groups based on their main actions or location.
    *   **Flexion**: Primarily by the **iliopsoas** (psoas major and iliacus), with assistance from the pectineus, rectus femoris, and sartorius.
    *   **Extension**: Primarily by the **gluteus maximus** and hamstring muscles (biceps femoris, semitendinosus, semimembranosus).
    *   **Medial (Internal) Rotation**: Primarily by the **tensor fascia latae** and fibers of the gluteus medius and minimus.
    *   **Lateral (External) Rotation**: Primarily by the **obturator muscles, quadratus femoris, and gemelli**, with assistance from the gluteus maximus, sartorius, and piriformis.
    *   **Adduction**: Primarily by the **adductor longus, brevis, and magnus**, with assistance from the gracilis and pectineus.
    *   **Abduction**: Primarily by the **gluteus medius and minimus**, with assistance from the tensor fascia latae and sartorius.

These muscular actions are critical for movements such as walking, standing, running, and stair climbing, and for maintaining balance and controlling leg alignment.

## Motions in the hip joint

The hip joint is a **true ball-and-socket, diarthrodial (synovial) joint** that connects the lower extremities to the axial skeleton, supporting body weight and allowing substantial mobility. It permits rotational motion in all directions.

### Types of Motion and Axes
The hip joint allows for movement in three major axes, all perpendicular to one another, with the center of the entire axis located at the femoral head.

*   **Flexion and Extension**: Occur around a **transverse axis** in the **sagittal plane**.
    *   **Flexion** describes the decrease in angle between parts. It involves moving the leg forward, lifting the knee toward the chest, bending down, or squatting. The femoral head spins posteriorly in flexion.
    *   **Extension** describes the increase in angle. It involves moving the leg backward. The femoral head spins anteriorly in extension.
*   **Medial (Internal) Rotation and Lateral (External) Rotation**: Occur along a **longitudinal (vertical) axis** in the **transverse (horizontal) plane**.
    *   **Medial rotation** is turning the thigh toward the midline of the body. It describes rotation towards the midline.
    *   **Lateral rotation** is turning the thigh away from the midline of the body. It describes rotation away from the midline.
*   **Abduction and Adduction**: Occur along a **sagittal (anterior-posterior) axis** in the **frontal plane**.
    *   **Abduction** is a movement away from the midline. It involves moving the leg out to the side.
    *   **Adduction** is a movement towards the midline. It involves moving one leg toward the other.

### Range of Motion (ROM)
The range of motion varies among individuals and activities.

*   **Flexion**: Typically 0° to approximately **120° with the knee flexed** (to release hamstring tension). It can be about 90° with the knee extended. Maximum hip motion occurs during flexion.
*   **Extension**: Ranges from about **5° to 30°** (hyperextension). The least movement at the hip occurs during extension.
*   **Abduction**: Typically 0° to **30°/45°**.
*   **Adduction**: Typically 0° to **20°/30°**.
*   **Internal Rotation**: Ranges from 0° to **30°/40°**. Less rotation occurs when the hip is extended due to soft tissue restriction.
*   **External Rotation**: Ranges from 0° to **40°/60°**.

Normal gait on level ground requires about 30° flexion, 10° hyperextension, 5° abduction, 5° adduction, and 5° of both medial and lateral rotation. Hip flexion of at least 120°, abduction of at least 20°, and internal/external rotation of at least 20° are necessary for normal daily activities. Activities like squatting, rising from a chair, or stair climbing require substantial ROM.

### Open-Chain vs. Closed-Chain Movements
Hip motions can be understood from two perspectives:

*   **Open-Chain Movements (Femur moving on Pelvis)**: Occur when the foot is not fixed to anything. For example, lifting the leg up or moving the leg while lying down.
    *   In open-chain hip flexion, if the hip moves past its normal range, the **pelvis may go into posterior tilt and the low back may round (lumbar flexion)** to provide more range.
    *   In open-chain hip extension, if the hip moves past its normal range, the **pelvis may go into anterior tilt and the spine into extension**.
    *   These coordinated movements are referred to as **lumbo-pelvic rhythm**.
*   **Closed-Chain Movements (Pelvis moving on Femur)**: Occur when the feet are fixed, such as standing with feet on the ground. In this scenario, the pelvis moves on the femoral head.
    *   **Posterior pelvic tilt** (squeezing glutes and tucking pelvis) creates relative **hip extension**.
    *   **Anterior pelvic tilt** creates relative **hip flexion**.
    *   **Lateral pelvic tilt** can lead to abduction or adduction in the hip joint. For instance, if the gluteus medius is weak, it can result in contralateral pelvic drop and adduction of the limb.
    *   **Pelvic rotation** (e.g., turning to the left while feet are fixed) can cause one leg to go into internal rotation and the other into external rotation.

### Muscular Contributions
A large number of powerful and well-balanced muscles surround the hip joint, contributing to its stability and enabling its wide range of motion. Approximately 22 to 27 separate musculotendinous units act on the hip joint. These muscles can be grouped by their primary actions:

*   **Flexion**: Primarily by the **iliopsoas** (psoas major and iliacus), assisted by the pectineus, rectus femoris, and sartorius.
*   **Extension**: Primarily by the **gluteus maximus** and hamstring muscles (biceps femoris, semitendinosus, semimembranosus).
*   **Medial (Internal) Rotation**: Primarily by the **tensor fascia latae** and fibers of the gluteus medius (anterior portion) and minimus. The piriformis can switch from lateral to medial rotator with increased hip flexion.
*   **Lateral (External) Rotation**: Primarily by the **obturator muscles, quadratus femoris, and gemelli**, with assistance from the gluteus maximus, sartorius, and piriformis.
*   **Adduction**: Primarily by the **adductor longus, brevis, and magnus**, with assistance from the gracilis and pectineus.
*   **Abduction**: Primarily by the **gluteus medius and minimus**, with assistance from the tensor fascia latae and sartorius.

The effectiveness of muscles can change based on the hip's position, a phenomenon known as "inversion of muscular action," which commonly manifests as a muscle's secondary function. For example, adductor muscles may act as hip flexors when the hip is neutral but as extensors when the hip is already flexed.

### Factors Affecting Motion
*   **Femoroacetabular Impingement (FAI)**: Morphologic alterations of the femur or acetabulum can lead to FAI, which limits hip ROM. Cam-type FAI involves an abnormal femoral head-neck junction impinging on the acetabular rim, especially during flexion and internal rotation. Pincer-type FAI results from acetabular overcoverage, pinching the labrum between the femur and acetabulum. FAI can cause reduced hip flexion and internal rotation.
*   **Developmental Dysplasia of the Hip (DDH)**: This condition involves a shallow acetabulum and/or poor femoral head coverage, which can also lead to hip pain and altered range of motion.
*   **Capsular Laxity**: Can lead to "microinstability" of the hip, increasing joint rotations and femoral head translations and potentially altering normal kinematics, possibly leading to degeneration.
*   **Age**: Hip extension ROM may diminish with age, while flexion remains relatively unchanged. The incidence of triangular labra decreases with increasing age, with rounded and irregular margins increasing.
*   **Surgical Approaches**: Different surgical approaches for total hip arthroplasty (THA) can affect hip kinematics. For example, the direct lateral approach can lead to the acetabulum being slightly more retroverted, while the posterior approach is opposite. Surgical release and repair of abductor musculature (lateral approach) or short external rotators (posterior approach) can impact gait mechanics and rotatory kinetics.

# Hip joint loading

Calculating the forces acting on the hip joint is a complex but crucial aspect of biomechanics, enabling a deeper understanding of joint function, injury, and disease. Researchers have developed various methods, ranging from direct measurements and simplified static analyses to sophisticated mathematical models and computational simulations.

### Early Methods: Direct Measurement and Simplified Static Analysis

1.  **Direct Measurement using Instrumented Implants:**
    The most accurate way to determine hip joint forces in living subjects is through **direct measurement using specially designed instrumented endoprostheses**. Pioneers like Rydell were among the first to attempt this, yielding force magnitudes in the hip of approximately 2.3 to 2.9 times body weight for single-leg stance and 1.6 to 3.3 times body weight for level walking. These prostheses are equipped with strain gauges and telemetric devices, allowing for real-time, three-dimensional measurement of the resultant hip joint force with high accuracy (within 1% at ~200 Hz). Such *in vivo* measurements have established reference libraries, like the Orthoload database, which are considered gold standards for validating computational models. However, direct measurements are technically complex and can only be performed in patients who have undergone total hip replacement.

2.  **Two-Dimensional (Static) Analysis / Free-Body Diagrams:**
    To estimate hip joint forces without invasive implants, simplified **two-dimensional analyses based on free-body diagrams** have been widely used, particularly for static conditions like standing. These models typically balance the moment arms of the body weight (acting as a load) and the abductor muscle forces (acting as effort) around the femoral head, which serves as the fulcrum.
    *   **Pauwels' Contributions:** Friedrich Pauwels published seminal work in 1935, analyzing forces on the hip during two-legged standing, one-legged standing, and walking. He calculated the eccentric force exerted by the upper body (head, trunk, upper limbs, and swinging leg) and the counterbalancing force from the abductor muscles. His models suggested that the resultant force across the hip joint could reach **more than four times body weight** during the single-support period of gait. He also showed that the **lever arm of the body weight is typically about three times as long as that of the abductor muscles**.
    *   **Limitations:** While useful for providing a rough estimate and clinical insights (e.g., how limping or using a cane reduces joint load), these models are simplified and often do not account for complex dynamics, the contribution of individual muscles, or three-dimensional force components.

### Intermediate Methods: Mathematical Modeling with Reduction and Optimization

1.  **Mathematical Models in General:**
    Mathematical models are crucial for estimating hip joint loads and forces in other anatomical structures (like muscles and ligaments) that cannot be directly measured easily. These models vary in complexity based on their underlying assumptions and the number of muscles and structures considered.

2.  **Reduction Methods:**
    Early mathematical models, like those introduced by Inman in 1947, often used a **reduction method** to simplify the problem of muscle redundancy. This involves reducing the number of independent muscle forces to be solved, often by lumping multiple muscles into a single representative force, usually in a two-dimensional framework. While simple and providing a rough estimation, this approach can lead to a loss of information about the function of individual muscles.

3.  **Optimization Approaches (Inverse Dynamics):**
    To overcome the indeterminacy of muscle forces (where more muscles cross a joint than degrees of freedom), **optimization procedures** are employed, typically within an **inverse dynamics approach**. In inverse dynamics, external forces and body motions are known, and the internal forces (e.g., muscle forces, joint reaction forces) are computed.
    *   **Optimization Criteria:** Various criteria are used to solve for the unknown muscle forces, such as minimizing the sum of muscle forces, muscle stresses (force per unit physiological cross-sectional area), or powers of these values (e.g., sum of muscle forces to the power of 'n'). Other criteria include minimizing muscle fatigue, joint contact forces, or signals from Golgi tendon organs.
    *   **Impact on Results:** While different optimization criteria can lead to varied distributions of individual muscle forces, the **magnitude and direction of the overall hip joint resultant force (R) itself do not significantly depend on the choice of the optimization criterion** for a given muscle model. However, alterations in the underlying **geometry of the muscle model** (e.g., attachment points) can significantly change the computed force R.
    *   **Model Complexity:** Models can range from including nine effective muscular units to **twenty-seven effective musculotendinous units**. These models define muscle attachments, physiological cross-sectional area (PCSA), maximal isometric force, and sometimes muscle receptor density.

### More Complex Musculoskeletal Analysis

1.  **Full-Body Musculoskeletal Modeling Packages:**
    Advances in computational power have led to the development of comprehensive **full-body musculoskeletal modeling packages** (e.g., AnyBody, OpenSim). These software simulate bones as rigid bodies connected by mechanical joints, with muscles functioning as actuators that provide the necessary torque for movement. They typically use an inverse dynamic approach to calculate internal forces from measured external forces and motions. These models allow for the evaluation of forces during daily activities, sports, and even surgical strategies without invasive *in vivo* experiments.

2.  **Subject-Specific Scalable Models:**
    The trend has shifted towards **individualized evaluation of joint biomechanics**. Subject-specific scalable lower extremity models (e.g., TLEM 2.0) are being developed to account for individual variations in muscle lever arms, leading to more accurate estimations of hip joint reaction forces.

3.  **Finite Element Analysis (FEA) / Finite Element Models (FEM):**
    **FEA is a powerful computational tool widely used in biomechanics** to model the response of the hip joint during various activities and to evaluate how the cartilage and other structures are affected by forces.
    *   **Model Creation:** FE models of the hip joint are often developed based on **CT (Computed Tomography) images** to create detailed 3D geometries of the bones (pelvis and femur) and cartilage. These models can then be subjected to simulated loads corresponding to various activities like one-leg stance or walking.
    *   **Stress Distribution:** FEA allows for the calculation of contact stress distribution within the hip joint's articular cartilage. The accuracy of these stress distributions depends on knowing the strain/stress relationships and loads of the biological structures at a microscopic level, which is challenging for complex biological structures like the hip.
    *   **Integration with Imaging and Machine Learning:** Research is also focused on **creating 3D FE models from 2D X-ray images using machine learning algorithms**. This approach aims to provide valuable biomechanical information, such as how pediatric hip deformities affect cartilage, without additional radiation exposure. This involves automatically detecting anatomical landmarks on X-ray images to reconstruct the femur and scale FE models to match individual subject sizes.
    *   **Applications:** FEA can help in understanding how forces affect different structures, which can inform treatment routes and interventions. It has been used to predict transchondral stress and strain in the human hip, and to validate predictions of cartilage contact pressure.

4.  **Discrete Element Analysis (DEA):**
    An alternative method for determining hip joint contact stress distribution is **discrete element analysis**, where the acetabulum and femur are modeled as arrays of smooth fine mesh elements connected by linear springs. Stress is then determined from the deformation of these springs.

### Factors Influencing Hip Joint Force Calculations

*   **Body Weight and Lever Arms:** The magnitude of hip joint forces depends critically on the lever arm ratio between the body weight moment arm and the abductor muscle moment arm. Increases in body weight significantly increase total compressive forces on the joint.
*   **Activities and Gait Speed:** Hip joint forces vary considerably with activity. They range from approximately 1.6 to 4.3 times body weight during walking, up to 5.5 times body weight during stair ascending/descending, and can reach up to 8 times body weight during activities like running or skiing. Faster gait speeds generally increase the load on the hip.
*   **Anatomical Variations:** Factors like a short femoral neck or a wide pelvis can lead to higher hip forces. Femoral anteversion also influences the "functional" femoral offset, affecting hip joint reaction forces.
*   **Pathologies:** Conditions such as developmental dysplasia of the hip (DDH), femoroacetabular impingement (FAI), and avascular necrosis (AVN) alter the hip's biomechanics, changing contact areas and increasing stress concentration, which can predispose to degeneration. For instance, dysplastic hips may experience contact forces as high as 260% of body weight during single-leg stance.
*   **Surgical Interventions:** Total hip arthroplasty (THA) influences all joint parameters including joint center, neck angle, offset, lever arms, and range of motion. The choice of surgical approach (e.g., lateral, posterior, anterior) can affect acetabular and femoral stem orientation, potentially impacting gait mechanics, rotatory kinetics, and dislocation rates. Surgical modifications like lateral displacement of the greater trochanter or varus/valgus osteotomies aim to optimize the abductor muscle lever arm and enlarge the weight-bearing surface to reduce joint pressure.
*   **External Support:** Using a cane or walking stick on the contralateral side can significantly reduce the joint reaction force on the painful hip by decreasing the gravitational moment arm and the required abductor muscle force.

## Diseases of the hip joint 

Hip replacement surgery, also known as total hip arthroplasty (THA), is a well-established surgical intervention commonly performed to address debilitating hip conditions that cause significant pain and functional impairments, ultimately reducing an individual's quality of life. While it was once primarily for older, low-demand patients, advancements mean it's increasingly performed on younger, more active individuals.

The main diseases and conditions leading to hip replacement include:

*   **Osteoarthritis (OA)**: This is the **most common reason** for total hip replacement, accounting for 75-80% of THAs. OA is typically an age-related "wear-and-tear" type of arthritis, often affecting individuals 50 years of age and older, sometimes with a family history. In OA, the cartilage cushioning the bones of the hip wears away, causing bones to rub against each other, leading to pain and stiffness. It can also be caused or accelerated by subtle irregularities in hip development during childhood. Radiographic findings often show asymmetrical joint space narrowing, especially in weight-bearing areas, along with osteophyte formation, subchondral bone sclerosis, and subchondral bone cysts.
*   **Avascular Necrosis (AVN) / Osteonecrosis (ON)**: This condition involves the decay of bone tissue in part of the femoral head due to a disrupted blood supply. It can result from injuries like dislocation or fracture that disrupt blood supply to the femoral head, or from certain diseases. AVN often presents in a younger patient population, typically between 35 and 50 years of age, and accounts for approximately 10% of annual THAs. In its terminal phase, the entire hip joint can be affected, with osteoarthritis superimposed on the deformed femoral head, making total hip replacement a common surgical solution.
*   **Rheumatoid Arthritis**: This is an autoimmune disease where the synovial membrane becomes inflamed and thick, leading to chronic inflammation that can damage cartilage and cause pain and stiffness. It is the most common type of inflammatory arthritis.
*   **Developmental Dysplasia of the Hip (DDH)**: This condition occurs when the hip joints do not fit together properly. It's a progressive condition that can cause structural damage to the hip and is a significant cause of early osteoarthritis, sometimes developing before 60 years of age. Approximately 20% to 50% of adults with hip dysplasia may develop early osteoarthritis by age 50. In dysplastic hips, contact between articular surfaces is concentrated on a small area of cartilage, and cadaveric studies have shown contact forces can be as high as 260% of body weight during single-leg stance. Untreated DDH can lead to gradual progression of functional disability and accelerated osteoarthritis.
*   **Hip Fractures**: Fractures, particularly of the femoral neck, can lead to the need for hip replacement, especially among elderly individuals with osteoporosis. These fractures often result in loss of balance and a fall. A hip fracture can disrupt the blood supply to the femoral head, predisposing it to avascular necrosis.
*   **Post-traumatic Arthritis**: This can develop after a serious hip injury or fracture, leading to cartilage damage, pain, and stiffness over time.
*   **Femoroacetabular Impingement Syndrome (FAI)**: FAI involves abnormal contact between the femur and acetabulum, contributing to labral tears and cartilage damage. It is often related to primary hip OA (coxarthrosis).
*   **Tumors**: Both benign and malignant bone tumors can necessitate hip replacement.
*   **Perthes Disease**: This is an aseptic femoral head necrosis primarily seen in children, which can lead to hip replacement later in life if it causes hip damage.
*   **Protrusio Acetabuli**: This condition, where the acetabulum indents into the pelvis, can also be an indication for THA.
*   **Ankylosing Spondylitis**: This inflammatory disease primarily affecting the spine can also involve the hip joints, leading to the need for replacement.
*   **Residual Arthritis from Slipped Capital Femoral Epiphysis**: A condition where the head of the femur slips off the growth plate, leading to arthritis that may eventually require THA.

Total hip arthroplasty aims to relieve pain, restore function, and improve mobility for patients suffering from these conditions.

## Total hip arthroplasty

Total hip arthroplasty (THA), also known as hip replacement surgery, is a widely performed and highly effective surgical intervention aimed at alleviating severe hip pain and restoring function and quality of life for individuals suffering from debilitating hip conditions. This procedure involves replacing damaged parts of the hip joint with artificial components to mimic the hip's natural function.

### Historical overview

The history of hip replacement dates back to the late 19th century. Early attempts at surgical solutions for hip pain included the experimental interposition of various tissues, such as fascia lata, skin, or even pig's bladder, between the articulating surfaces of the hip. In 1891, Professor Themistocles Glück in Germany notably used ivory to replace femoral heads damaged by tuberculosis. In the early 20th century, artificial hip socket components were installed, with Smith-Peterson installing one in 1923, and Wiles in England designing a stainless steel total hip in 1938, though these early designs often did not function well. The 1960s marked a significant turning point with the introduction of the cemented McKee-Farrar THA, which became the first widely adopted successful total hip arthroplasty. Around the same time, Sir John Charnley, often recognized as the "father of modern total hip replacement," revolutionized the procedure with his "low-friction arthroplasty" concept, initially using a Teflon acetabular component and a stainless steel femoral endoprosthesis, both cemented to bone. Despite early issues with excessive wear of Teflon, Charnley's subsequent use of high-density polyethylene in 1962 laid the groundwork for modern THA.

### Components of hip arthroplasty

Modern hip arthroplasty prostheses consist of several key components designed to replace the damaged bone and cartilage of the hip joint. These artificial parts are typically made from a combination of **metal, plastic (polyethylene), or ceramic**. The prosthesis generally includes:
*   **Femoral Head:** A metal or ceramic ball that replaces the damaged femoral head.
*   **Femoral Stem:** A metal stem that is placed into the hollow center of the femur. This stem can be either **cemented** into the bone or designed for a "press fit," allowing for bone ingrowth.
*   **Acetabular Component:** A metal socket that replaces the damaged cartilage surface of the acetabulum. This component may be held in place with screws or cement, or it can be a "press-fit" type with a roughened surface to encourage bone ingrowth.
*   **Liner/Spacer:** A plastic (polyethylene), ceramic, or metal spacer that is inserted between the new ball and the socket, creating a smooth gliding surface for articulation. Ceramics like alumina and zirconia are favored for their exceptional wear resistance, biocompatibility, and low wear rates, which help minimize bone dissolution caused by wear particles.

There are several types of hip arthroplasty, each with distinct characteristics and indications:
*   **Total Hip Arthroplasty (THA)**: This is the most common type of hip replacement. It involves replacing **both the femoral head (ball) and the acetabulum (socket)** with prosthetic components. THA is typically recommended when both sides of the hip joint are affected by disease, such as in cases of **osteoarthritis (OA)**, which is the most frequent reason for the procedure.
*   **Partial Hip Replacement (Hemiarthroplasty)**: This procedure replaces **only the femoral head** and neck, leaving the acetabulum intact. Hemiarthroplasty is less common for general hip replacement and is usually reserved for specific situations like certain types of **hip fractures**, particularly of the femoral neck, or for tumor removal. It is primarily recommended for elderly or frail patients due to concerns about the prosthesis loosening or eroding the native acetabulum over time.
*   **Hip Resurfacing Arthroplasty (HRA)**: This is an alternative to conventional THA that aims to **preserve more of the patient's original bone**. In HRA, the arthritic surfaces of the joint are replaced, but less bone is removed from the femur compared to a total hip replacement. This procedure typically involves a large femoral head articulation, which can result in a biomechanically more stable joint and potentially lower dislocation rates. HRA is generally considered ideal for **young, active adults** with good bone quality and hip morphology. However, certain contraindications exist, including proximal femoral osteoporosis, tumors, inflammatory arthritis, poor bone stock, and acetabular dysplasia. While early metal-on-metal HRA designs faced challenges with high revision rates due to metal debris, modern advancements include ceramic-on-ceramic or metal-on-polyethylene bearing surfaces to address these issues.

### Fixation method

The choice of fixation method for the prosthetic components in hip arthroplasty is a crucial decision influenced by various factors, including patient age, bone quality, activity level, and surgeon preference. Hip arthroplasty components are primarily fixed into the bone using either **cemented** or **uncemented (biological ingrowth)** methods, with hybrid approaches also commonly employed.

**Cemented fixation** involves using polymethyl methacrylate cement to bond the prosthetic component to the bone. Historically, Sir John Charnley's "low-friction arthroplasty" laid the groundwork for modern cemented THA using high-density polyethylene [previous response]. The success of cementing relies heavily on the surgical technique, with continuous improvements over decades including vacuum-mixed cements, medullary canal plugs, centralizing elements, and jet-lavage to clean trabecular bone structures. Cemented prostheses have shown statistically the **best results in terms of the overall THA population**, particularly for older patients. For individuals over 75 years, cemented total hip arthroplasty is strictly recommended. However, cement performs better under compression than shearing forces, which historically led to increased failure rates for cemented acetabular cups. Loosening in cemented prostheses is often indicated by a lucency wider than 2 mm or progression of lucencies at the bone-cement interface on radiographs. National Joint Registry data for England and Wales indicated that cemented THAs had revision rates of 0.6% at one year, 1.4% at three years, and 2.0% at five years, which are generally lower than uncemented options.

**Uncemented (biological ingrowth)** fixation relies on the ingrowth of the patient's bone directly onto or into a porous or roughened surface of the implant, creating a solid biological bond. This "press-fit" method requires the implant to be impacted into a slightly smaller cavity to achieve initial stability. Successful bone ingrowth is dependent on the quality of the initial fixation achieved by the surgeon, the characteristics of the implant surface, and the quality of the reamed bone bed, with micromotion in the early postoperative period needing to remain below a critical threshold. Uncemented fixation is commonly considered for **younger and more active patients** with good bone quality, and it is predominantly used for both femoral and acetabular components in cases of avascular necrosis. However, uncemented THAs have shown higher revision rates compared to cemented ones in some registries, with 1.3% at one year, 2.5% at three years, and 3.4% at five years.

**Hybrid fixation** combines an uncemented acetabular component with a cemented femoral stem. This approach leverages the strengths of both methods, aiming to provide good stability for the acetabular cup and reliable fixation for the femoral stem. Hybrid prostheses are increasingly popular, with revision rates catching up to cemented options. For patients aged between 60 and 70 who are borderline active, hybrid total hip arthroplasty may be considered. National Joint Registry data reported hybrid THA revision rates of 0.9% at one year, 1.8% at three years, and 2.7% at five years. The decision on which fixation method to use depends on the patient's individual circumstances, with age, bone density, and expected activity level being key considerations for optimizing long-term outcomes and minimizing complications.

## Failures of THA

Total hip arthroplasty (THA), despite being a highly successful surgical intervention, is associated with various potential failures and complications that can occur due to a range of factors including implant design, materials, surgical technique, and patient characteristics. Understanding these potential issues is crucial for optimizing long-term outcomes.

Possible failures and their reasons include:

*   **Loosening and Osteolysis:**
    *   **Aseptic Loosening** is one of the most common causes for revision surgery, accounting for approximately 19.7% of revisions in some reports and 30% in others from national registries. It is also the most common problem seen with **cemented prostheses**. It can be indicated radiographically by a lucency wider than 2 mm or progression of lucencies at the bone-cement interface. Cemented acetabular cups, in particular, historically had increased failure rates because cement performs better under compression than shearing forces [previous response]. Poor cementing technique, such as a lack of cement mantle at the stem tip, can lead to early failure.
    *   **Osteolysis**, the loss of bone, is often linked to long-term problems with hip replacements. It is primarily caused by the body's reaction to **polyethylene wear debris**, fine particles of plastic wearing off the liner, which triggers an inflammatory process and bone resorption. This can lead to loosening of implants and even fractures around them. Early cemented components also showed osteolysis, initially thought to be "cement disease," but later attributed to polyethylene wear debris.
    *   For **uncemented (biological ingrowth) fixation**, micromotion above a critical threshold in the early postoperative period can prevent successful bone ingrowth, leading to failure. Too much oversizing during "press-fitting" can cause bone fractures or fissures and small contact areas, while too little can lead to implant movement.
    *   **Femoral neck narrowing and scalloping** can also occur, secondary to stress shielding, loosening, or an inflammatory response, including metal hypersensitivity.

*   **Dislocation:**
    *   **Dislocation**, where the femoral head comes out of the socket, is one of the most common complications, accounting for approximately 28% of revisions in some registries. About 70% of THA dislocations occur within the first month after surgery, primarily due to incomplete scar formation and relaxed soft tissues.
    *   Key risk factors include **component malpositioning** (excessive anteversion can lead to anterior dislocation, while excessive retroversion can lead to posterior dislocation). Component orientation (inclination and anteversion) significantly influences dislocation risk.
    *   Other factors contributing to dislocation risk include prior hip surgery (most significant independent risk factor), elderly age (over 70 years), neuromuscular conditions (e.g., Parkinson's disease), and drug/alcohol abuse. Preservation of gluteal muscles and restoration of leg length and femoral offset are important for reducing dislocation risk. A femoral offset of less than 33 mm is associated with hip dislocation.

*   **Infection:**
    *   **Infection** can occur superficially in the wound or deep around the prosthesis, happening days, weeks, or even years after surgery. It is one of the most common causes for revision of a THA, with an incidence of about 1% to 2% in primary THA.
    *   Risk factors include obesity, diabetes, smoking, immunosuppressive medications or diseases, history of infection, previous hip surgery, IV drug use/abuse, alcohol abuse, poor oral hygiene, chronic renal/liver disease, malnutrition, and HIV (CD4 counts less than 400).

*   **Fractures:**
    *   **Bone fractures** around the prosthesis (periprosthetic fractures) are a known complication, with an increasing incidence, particularly in younger patient populations. They are also common in very elderly patients with osteoporosis and trivial trauma.
    *   **Intraoperative fractures** can occur in either the acetabulum or the femur during the procedure. Acetabular fractures (0.4% incidence with press-fit components) often occur during component impaction, with risk factors including under-reaming, poor patient bone quality, and dysplastic conditions. Femur fractures (up to 5% incidence) can result from technical errors, use of press-fit implants, poor patient bone quality, or revision surgery.
    *   Specific to **Hip Resurfacing Arthroplasty (HRA)**, **femoral neck fracture** is the most common complication (12% of cases), more common in females, and may be related to osteoporosis or poor surgical technique (e.g., notching of the femoral neck, varus positioning).

*   **Bearing Surface-Specific Issues:**
    *   **Metal-on-metal (MoM) bearings** have faced significant issues, with higher failure rates compared to metal-on-plastic or ceramic-on-ceramic (6.2% vs. 1.7% and 2.3% at 5 years). Failures are linked to the release of minute metallic particles or metal ions, causing pain, disability, and adverse local tissue reactions such as **metallosis** and **pseudotumors**. Larger MoM head sizes have been associated with increased failure rates. Malpositioning, particularly cup inclinations above 50° and/or anteversion above 15°, can lead to "run-away wear".
    *   **Ceramic-on-ceramic (CoC) bearings** are known for low wear rates and debris. However, challenges include **ceramic fragmentation** (catastrophic failure can occur in about 2% of implants) and **abnormal joint noise** ("squeaking"). Increased friction due to rim loading or fluid film breakdown can lead to stripe wear, vibrations, and high moments at the implant-bone interface, causing fixation problems. CoC bearings are more sensitive to precise positioning.
    *   **Polyethylene (PE) bearings**, while widely used, can lead to **polyethylene wear debris** which causes periprosthetic osteolysis and aseptic loosening. Highly cross-linked polyethylene was developed to reduce wear but may be less tough and prone to cracking.

*   **Other Complications:**
    *   **Blood clots** (deep vein thrombosis in legs/pelvis, pulmonary embolism in lungs) are common complications, and can be life-threatening if they travel to the lungs.
    *   **Nerve damage** (e.g., sciatic or lateral femoral cutaneous nerve palsy) can cause numbness, weakness, and pain, though often transient.
    *   **Limb length discrepancy** can occur, where one leg becomes longer or shorter, sometimes due to muscle contracture.
    *   **Stiffness** in the joint is possible.
    *   **Heterotopic ossification** (abnormal bone formation around the hip) is a recognized complication.
    *   **Chronic pain** can persist after surgery, potentially from iliopsoas impingement, trochanteric bursitis, or nerve injury.
    *   **Adverse tissue reactions** can occur, including granulomatous disease, related to prosthesis failure, particularly type VI hypersensitivity response to metal ions.
    *   **Technical difficulties** during surgery, such as femoral canal perforation (6.8% in some sickle cell anemia cases) or difficult prosthesis reduction, can lead to complications.

Overall, the long-term success of THA is dependent on addressing factors such as implant stability (initial and long-term), wear particle generation, anatomical restoration (offset, component orientation), surgical precision, and careful patient selection, particularly for younger, more active individuals.
